```python
def faradaycal_new(I,Q,U,V,freq_test,trial_RM,trial_phi,width_native,t_samp,plot=False,datadir=".",calstr="",label="",n_f=1,n_t=1,show=False):
    #Get wavelength axis
    c = (3e8) #m/s
    wav = c/(freq_test[0]*(1e6))#wav_test[0]
    wavall = np.array([wav_test[0]]*np.shape(I)[1]).transpose()
    
    peak,timestart,timestop = dsapol.find_peak(I,width_native,t_samp)
    #print(timestop-timestart)

    #Calculate polarization
    P = Q + 1j*U

    #SNR matrix
    SNRs = np.zeros((len(trial_RM),len(trial_phi)))

    for i in range(len(trial_RM)):
        for j in range(len(trial_phi)):
            RM_i = trial_RM[i]
            phi_j = trial_phi[j]
            """
            P_trial = np.zeros(np.shape(P),dtype=complex) #P*np.exp(-1j*((2*RM_i*((wav)**2)) + phi_j))
            #print(np.shape(P))
            for k in range(np.shape(P)[1]):
                #print(np.shape(P_trial[k,:]))
                P_trial[:,k] = P[:,k]*np.exp(-1j*((2*RM_i*((wav)**2)) + phi_j))
            """
            P_trial = P*np.exp(-1j*((2*RM_i*((wavall)**2)) + phi_j))
            Q_trial_t = np.mean(np.real(P_trial),axis=0)
            U_trial_t = np.mean(np.imag(P_trial),axis=0)
            
            #L_trial = np.sqrt(np.real(P_trial)**2 + np.imag(P_trial)**2)
            #L_trial_t = np.mean(L_trial,axis=0)
            
            L_trial_t = np.sqrt(Q_trial_t**2 + U_trial_t**2)
            """
            plt.figure()
            plt.plot(np.real(np.mean(P_trial,axis=0)))
            plt.plot(np.imag(np.mean(P_trial,axis=0)))
            plt.plot(L_trial_t)
            plt.xlim(timestart,timestop)
            plt.show()
            """
            
            sig = np.mean(L_trial_t[timestart:timestop])#np.abs(np.mean(np.mean(P_trial,axis=0)[timestart:timestop]))#
            
            L_trial_offp = np.concatenate([L_trial_t[:timestart],L_trial_t[timestop:]])
            L_trial_offp = L_trial_offp[:len(L_trial_offp) - (len(L_trial_offp)%(timestop-timestart))]
            
            L_trial_binned_offp = np.mean(L_trial_offp.reshape(len(L_trial_offp)//(timestop-timestart),timestop-timestart))
            #print(L_trial_offp)
            #print(L_trial_binned_offp)
            
            """
            L_trial_binned_offp = np.zeros(len(L_trial_t)-(timestop-timestart))
            width = timestop-timestart
            for k in range((len(L_trial_t)-(timestop-timestart))//(timestop-timestart)):
                if (k+1)*width >= len(L_trial_offp):
                    L_trial_binned_offp[k] = np.mean(L_trial_offp[k*width:])
                else:
                    L_trial_binned_offp[k] = np.mean(L_trial_offp[k*width:(k+1)*width])
            """
            #binned_off_pulse =  #dsapol.avg_time(np.concatenate([L_trial_t[:timestart],L_trial_t[timestop:]])[:4096],timestop-timestart)
            noise = np.sqrt(np.mean(L_trial_binned_offp**2))
            #print(sig,noise)
            SNRs[i,j] = sig/noise#np.abs(np.sum(P_trial))

    
    (max_RM_idx,max_phi_idx) = np.unravel_index(np.argmax(SNRs),np.shape(SNRs))
    lower = trial_RM[np.argmin(np.abs((SNRs[:,0])-(np.max(SNRs[:,0]) - 1)))]
    upper = trial_RM[max_RM_idx]-trial_RM[np.argmin(np.abs((SNRs[:,0])-(np.max(SNRs[:,0]) - 1)))] + trial_RM[max_RM_idx]
    RMerr = (upper-lower)/2
    if plot:

        f=plt.figure(figsize=(12,6))
        plt.grid()
        plt.plot(trial_RM,SNRs)
        plt.axvline(lower,color="red")
        plt.axvline(upper,color="red")
        plt.xlabel("Trial RM")
        plt.ylabel("SNR")
        plt.title(label)
        #plt.savefig(datadir + label + "_faraday1D_" + calstr + str(n_f) + "_binned" + ext)
        if show:
            plt.show()
    return (trial_RM[max_RM_idx],trial_phi[max_phi_idx],SNRs[:,0],RMerr)
    
```

```python
def get_error(Q_f_cal,U_f_cal,wav_test,RM,phi=0):
    P_f_cal_RM = (Q_f_cal + 1j*U_f_cal)*np.exp(-1j*((2*(RM-100)*((np.array(wav_test[0]))**2)) + phi))

    P_mag = np.abs(np.sum(P_f_cal_RM))
    #print(P_mag)
    sigma_Q = np.sqrt(np.mean(np.real(P_f_cal_RM)**2))
    sigma_U = np.sqrt(np.mean(np.imag(P_f_cal_RM)**2))
    #print(sigma_Q,sigma_U)
    sigma = sigma_Q
    SNR = P_mag*np.sqrt((len(wav_test[0]) -2))/sigma
    #print(SNR)
    sigma_lambda = np.sqrt(np.sum(np.array(wav_test[0])**4)/(len(wav_test[0]) - 1))
    #print(sigma_lambda)
    sigma_chi = 0.5*sigma/P_mag
    #print(sigma_chi)
    sigma_RM = sigma_chi/(sigma_lambda*np.sqrt(len(wav_test[0] )- 2))
    print(sigma_RM)
    #print(len(wav_test[0]))
    #print(np.sum(np.array(wav_test[0])**4))
    method1HWHM = sigma_RM
    return method1HWHM
```
