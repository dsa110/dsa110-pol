```python
gain_dir = '/home/ubuntu/sherman/scratch_weights_update_2022-06-03/3C48_test/'
gain_source_name = "3C48"
gain_obs_names = ["ane"]
n_t = 1
n_f = 32
nsamps = 5120
deg = 10
suffix = "_dev"

phase_dir ="/home/ubuntu/sherman/scratch_weights_update_2022-06-03/3C286_test/"
phase_source_name = "3C286"
phase_obs_names = ["jqc"]
ratio,ratio_fit_params = dsapol.gaincal_full(datadir=gain_dir,source_name=gain_source_name,obs_names=gain_obs_names,n_t=n_t,n_f=n_f,nsamps=nsamps,deg=deg,suffix=suffix,average=True,plot=True)

ratio_fit = np.zeros(np.shape(freq_test[0]))
for i in range(deg+1):
    ratio_fit += ratio_fit_params[i]*(freq_test[0]**(deg-i))
ratio_use = ratio_fit
plt.figure()
plt.plot(freq_test[0],ratio)
plt.plot(freq_test[0],ratio_use)

phase_diff,phase_fit_params = dsapol.phasecal_full(datadir=phase_dir,source_name=phase_source_name,obs_names=phase_obs_names,n_t=n_t,n_f=n_f,nsamps=nsamps,deg=deg,suffix=suffix,average=True,plot=True)

phase_fit = np.zeros(np.shape(freq_test[0]))
for i in range(deg+1):
    phase_fit += phase_fit_params[i]*(freq_test[0]**(deg-i))
phase_use = phase_fit

plt.figure()
plt.plot(freq_test[0],phase_diff)
plt.plot(freq_test[0],phase_use)

```

```python
(I_f_cal_RM,Q_f_cal_RM,U_f_cal_RM,V_f_cal_RM) = dsapol.get_stokes_vs_freq(I_cal_RM,Q_cal_RM,U_cal_RM,V_cal_RM,1,fobj.header.tsamp,32,freq_test,plot=True,datadir=datadir,label=label,calstr=calstr,show=True)
```

```python
P_f_cal_RM = Q_f_cal_RM + 1j*U_f_cal_RM
#P_f_cal_RM = (Q_f_cal + 1j*U_f_cal)*np.exp(-1j*((2*(RM-100)*((np.array(wav_test[0]))**2)) + phi))

P_mag = np.abs(np.sum(P_f_cal_RM))
#print(P_mag)
sigma_Q = np.sqrt(np.mean(Q_f_cal_RM**2))
sigma_U = np.sqrt(np.mean(U_f_cal_RM**2))
#print(sigma_Q,sigma_U)
sigma = sigma_Q
SNR = P_mag*np.sqrt((len(wav_test[0]) -2))/sigma
#print(SNR)
sigma_lambda = np.sqrt(np.sum(np.array(wav_test[0])**4)/(len(wav_test[0]) - 1))
#print(sigma_lambda)
sigma_chi = 0.5*sigma/P_mag
#print(sigma_chi)
sigma_RM = sigma_chi/(sigma_lambda*np.sqrt(len(wav_test[0] )- 2))
print(sigma_RM)
#print(len(wav_test[0]))
#print(np.sum(np.array(wav_test[0])**4))
method1HWHM = sigma_RM
```

```python
print(np.pi/((wav_test[0][-1]**2-wav_test[0][0]**2)))
method2HWHM= np.pi/((wav_test[0][-1]**2-wav_test[0][0]**2))/2
```

```python
fit_sample_RM = trial_RM[np.argmax(SNRs_cal)-100:np.argmax(SNRs_cal)+100]
fit_sample_SNRs = SNRs_cal[np.argmax(SNRs_cal)-100:np.argmax(SNRs_cal)+100]
plot_sample_RM = trial_RM[np.argmax(SNRs_cal)-200:np.argmax(SNRs_cal)+200]
plot_sample_SNRs = SNRs_cal[np.argmax(SNRs_cal)-200:np.argmax(SNRs_cal)+200]
plt.figure(figsize=(12,6))
plt.plot(plot_sample_RM,plot_sample_SNRs,'o',label="RM spectrum")
#plt.show()

popt = np.polyfit(fit_sample_RM,fit_sample_SNRs,2)
fit_RM = np.linspace(trial_RM[np.argmax(SNRs_cal)-100],trial_RM[np.argmax(SNRs_cal)+100],10000)
fit = np.zeros(10000)
for i in range(len(popt)):
    fit += popt[i]*(fit_RM**(len(popt)-1-i))
plt.plot(fit_RM,fit,label="Parabolic Fit")

def SNR_fit(RM,popt):
    fit = 0
    for i in range(len(popt)):
        fit += popt[i]*(RM**(len(popt)-1-i))
    return fit

HWHM_RM = RM_cal
step = 0.001
while SNR_fit(HWHM_RM,popt) > np.max(SNRs_cal)/2:
    HWHM_RM += step
print( np.max(SNRs_cal)/2,SNR_fit(HWHM_RM,popt),HWHM_RM)

plt.axvline(RM_cal-method1HWHM,color="purple")
plt.axvline(RM_cal+method1HWHM,color="purple",label="Method 1 (Large SNR RM Fit Error, " + str(np.around(method1HWHM,2)) + ")")

plt.axvline(RM_cal-method2HWHM,color="green")
plt.axvline(RM_cal+method2HWHM,color="green",label="Method 2 (HWHM estimate from " + r'$\Delta \lambda^2$ '  + str(np.around(method2HWHM,2)) + ")")


plt.axvline(HWHM_RM,color="red")

HWHM = (HWHM_RM - RM_cal)
plt.axvline(RM_cal-HWHM,label="Method 3 (Parabolic Fit HWHM, " + str(np.around(HWHM,2)) + ")",color="red")




plt.grid()
plt.xlabel("Trial RM")
plt.ylabel("SNR")
plt.legend()
plt.show()
print(HWHM)
```
